2. Linearize the following function at $x_0$: $f(x) = 2x^2 - x^3$  

- Write the equation of a line tangent at $ùë•_0 = 4$:

```python
import matplotlib.pyplot as plt
equation = 2*x**2 + x**3
derivative = sp.diff(equation, x)
display(derivative) # display

x0 = 4
display(derivative.subs(x, x0))
derivtive_at_x0 = derivative.subs(x, x0) * (x-x0) + equation.subs(x,x0)

sp.plot(equation, derivtive_at_x0, (x, -6, 6), ylim=(-200, 200))
plt.show()
```
